### This notebooks assumes we have the evaluation results CSV files stored under datasets/evaluation/results location which is obtained from running Experiments_Hugging_Face_API.ipynb Notebook

```python
import pandas as pd
import os
```

### Combining Individual Output files into one

```python
os.chdir('datasets/evaluation/results/files')
```

```python
os.listdir()
```

```python
files = [file for file in os.listdir()]
```

```python
df = pd.concat(map(pd.read_csv, files), ignore_index=True)
```

```python
df.tail(10)
```

```python
df.info()
```

```python
#Filling NA values with 0
df=df.fillna(0)
```

```python
df.info()
```

```python
#Dropping columns
df_new=df.drop(columns=['question','answer','contexts','ground_truth'],axis=1)
```

```python
df_new.head()
```

```python
df_new['llm'].unique()
```

```python
#Taking average of the scores for each combinations
df_new=df_new.groupby(['llm','embedding_model'], as_index=False).mean()
```

```python
df_new.head()
```

```python
df_new.tail()
```

## Finding Harmonic mean as RAGAS Score

```python
from scipy.stats import hmean
#Finding ragas score via harmonic mean
h_list=[]
for index,row in df_new.iterrows():
   h_mean=hmean([row['context_precision'], row['faithfulness'],row['answer_relevancy'], row['context_recall']])
   h_list.append(h_mean)
```

```python
df_new['ragas_score']=h_list
```

```python
df_eval_results=df_new.round(3)
df_eval_results=df_new.sort_values('ragas_score',ascending=False)
```

```python
#Saving the final results 
df_eval_results.to_csv("../eval_results.csv",index=False)
```

```python
# Calculate the average metrics for each LLM and embedding model combination
average_metrics = df_new.groupby(['llm', 'embedding_model']).mean().reset_index()

# Create bar plots for each metric
metrics = ['context_precision', 'faithfulness', 'answer_relevancy', 'context_recall', 'answer_similarity', 'answer_correctness', 'ragas_score']

fig, axs = plt.subplots(len(metrics), 1, figsize=(10, 35))

for i, metric in enumerate(metrics):
    sns.barplot(x='llm', y=metric, hue='embedding_model', data=average_metrics, ax=axs[i])
    axs[i].set_title(f'Average {metric.replace("_", " ").title()} by LLM and Embedding Model')
    axs[i].set_xlabel('LLM')
    axs[i].set_ylabel(metric.replace("_", " ").title())
    axs[i].legend(loc='upper right')

plt.tight_layout()
plt.show()
```
